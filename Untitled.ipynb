```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('modified.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train ,y_test = train_test_split(df.drop('classification' , axis = 1) , df['classification'] , test_size = 0.2)
```

```python
y_test
```

```python
import numpy as np

# Clean numeric columns
cols1 = ['age' , 'bp' , 'sg' , 'al' , 'su' , 'bgr' , 'bu' , 'sc' , 
         'sod' , 'pot' , 'hemo' , 'pcv' , 'wc' , 'rc']
# Clean only cols1
for col in cols1:
    X_train[col] = (
        X_train[col]
        .astype(str)                                # ensure string
        .str.replace(r'[\t?]', '', regex=True)      # remove tabs and '?'
        .str.strip()                                # strip spaces
    )
    X_train[col] = pd.to_numeric(X_train[col], errors='coerce')  # convert → numeric

# Apply imputer only on cols1
si = SimpleImputer(strategy="mean")
X_train[cols1] = si.fit_transform(X_train[cols1])
```

```python
# Define categorical columns and mapping dictionary
categorical_columns = ['htn' , 'dm' , 'cad' , 'pe' , 'ane' , 'rbc' , 'pc' , 'pcc' , 'ba' ]
cols = ['htn', 'dm' , 'cad' , 'pe' , 'ane']
cats = [['no' , 'yes'] , ['no' , 'yes'] , ['no' , 'yes'] , ['no' , 'yes'] , ['no' , 'yes']]
```

```python
#Applying One Hot Encoding on sex and Embarked column
# One Hot Encoders
ohe_rbc = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_pc = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_pcc = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_ba = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
X_train_rbc = ohe_rbc.fit_transform(X_train[['rbc']])
X_train_pc = ohe_pc.fit_transform(X_train[['pc']])
X_train_pcc = ohe_pcc.fit_transform(X_train[['pcc']])
X_train_ba = ohe_ba.fit_transform(X_train[['ba']])

#Transform Test
X_test_rbc = ohe_rbc.transform(X_test[['rbc']])
X_test_pc = ohe_pc.transform(X_test[['pc']])
X_test_pcc = ohe_pcc.transform(X_test[['pcc']])
X_test_ba = ohe_ba.transform(X_test[['ba']])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
```

```python
y_train = le.fit_transform(y_train)
y_train_le = y_train.reshape(-1, 1)
```

```python
y_test = le.fit_transform(y_test)
y_test_le = y_test.reshape(-1, 1)
```

```python
y_test
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
# Define categorical columns and mappings
cols2 = ['htn', 'dm', 'cad', 'pe', 'ane']
cats = [['no', 'yes'], ['no', 'yes'], ['no', 'yes'], ['no', 'yes'], ['no', 'yes']]
for col in cols2:  # cols = ['htn', 'dm', 'cad', 'pe', 'ane']
    X_train[col] = X_train[col].astype(str).str.replace('\t', '').str.strip().str.lower()
    X_test[col]  = X_test[col].astype(str).str.replace('\t', '').str.strip().str.lower()

# Apply ordinal encoding
encoder = OrdinalEncoder(categories=cats)
X_train[cols2] = encoder.fit_transform(X_train[cols2])
X_test[cols2]  = encoder.transform(X_test[cols2])
```

```python
X_train_transform = np.concatenate((X_train[cols1] , X_train_rbc , X_train_pc , X_train_pcc , X_train_ba , y_train_le , X_train[cols2]) , axis = 1)
```

```python
X_test_transform = np.concatenate((X_test[cols1] , X_test_rbc , X_test_pc , X_test_pcc , X_test_ba , y_test_le , X_test[cols2]) , axis = 1)
```

```python
X_train_transform.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transform , y_train_le)
```

```python
y_pred = clf.predict(X_test_transform)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```
